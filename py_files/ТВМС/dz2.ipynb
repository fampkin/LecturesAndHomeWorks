```python
from matplotlib import rcParams
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import locale
import random
import numpy as np
from scipy.stats import *
locale.setlocale(locale.LC_NUMERIC, 'russian')
plt.rcParams['axes.formatter.use_locale'] = True


plt.rcParams['font.size'] = 36
plt.rcParams['mathtext.fontset'] = 'cm'
```

```python
def condition(x,y):
    if x<=y<=x+1/6 or y<=x<=y+1/5:
        return True 
    else:
        return False
rcParams['figure.figsize'] = (5, 5)
rcParams['figure.dpi'] = 200
fix, ax = plt.subplots(figsize =(5,5))
plt.tick_params(labelsize=16)

ax.yaxis.set_major_locator(ticker.MultipleLocator(0.1)) 
plt.xlim(0,1) 
plt.ylim(0,1)
ax.set_aspect('equal', adjustable='box')

N = 15000
count = 0
for i in range(N):
   x = random.uniform(0,1)
   y = random.uniform(0,1)
   if condition(x,y): # должно быть условие для события A, что встреча состоится 
        plt.plot(x,y,marker='.', markersize=1,color ='r')
        count+=1
   else:
      plt.plot(x,y,marker='.', markersize=1,color ='gray')
plt.show()
#fig.savefig("Sol_4_1.pdf", bbox_inches='tight')
#fig.savefig("Sol_3_1.png", bbox_inches='tight') 
Pstat=count/N
Pstat
```

```python
#стас вер
N=1000000
count=0
x=[]
y=[]
for i in range(N+1):
    x.append(i+1)
    x1 = random.uniform(0,1) # Можно и так, но лучше через метод rvs для равномерного распрледеления из U\t 
    y1 = random.uniform(0,1)
    if condition(x1,y1): # должно быть условие для события А, что встреча состоится
        #plt.plot(x,y,'ro',color="red")
        #plt.plot(x,y,marker='.', markersize=1,color ='red')
        count+=1
    y.append(count/(i+1))
#plt.show()
#fig.savefig("Sol_3_1.png", bbox_inches='tight')
Pstat=count/N
prob=Pstat
Pstat
```

```python
rcParams['figure.figsize'] = (15, 10)
rcParams['figure.dpi'] = 300

fig, ax =plt.subplots(figsize=(30, 10))
plt.tick_params(labelsize = 36)
#ax.xaxis.set_major_Locator(ticker.MultipleLocator(200000))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.005))
ax.xaxis.set_major_locator(ticker.MaxNLocator(15))
ax.xaxis.set_minor_locator(ticker.MaxNLocator(8))
ax.yaxis.set_major_locator(ticker.MaxNLocator(5))
ax.yaxis.set_minor_locator(ticker.MaxNLocator(8))
plt.ylim(Pstat-0.005,Pstat+0.005)
plt.xlim(0.1,N)
#fig=plt.figure()
#plt.tick_params(Labelsize = 20)
plt.grid(color='DarkTurquoise', alpha=0.75,linestyle=':', linewidth=1)
plt.plot(x,y,color='b',lw=3)
plt.plot(x,[Pstat]*len(x),color='r',lw=2)
plt.show()
fig.savefig("Sol_2_1stat.png", bbox_inches='tight')
```

```python
def condition(x,y):
    if x+1/4>1:
        if x<=y<=1:
            return True
    if y+1/4>1:
        if y<=x<=1:
            return True

    if x<=y<=x+1/4 or y<=x<=y+1/4:
        return True   
 
    return False
```

```python
rcParams['figure.figsize'] = (5, 5)
rcParams['figure.dpi'] = 200
fig, ax =plt.subplots(figsize=(5, 5))
plt.tick_params(labelsize = 16)
#plt.grid(color='DarkTurquoise', alpha=0.75, LinestyLe=':', Linewidth=1) ax.xaxis.set_major_locator(ticker.MultipleLocator(0.1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.1)) 
plt.xlim(0,1) 
plt.ylim(0,1)
ax.set_aspect('equal', adjustable='box')
#ax.xaxis.set_major_Locator(ticker.MaxNLocator(10)) #ax.xaxis.set_minor_Locator(ticker.MaxNLocator(8))
#plt.axis('equal')


N=15000
count=0
for i in range(N):
    x = random.uniform(0,1)
    y = random.uniform(0,1)
    
    if condition(x,y): 
        plt.plot(x,y,marker='.', markersize=1,color ='red')
        count+=1
    else:
        plt.plot(x,y,marker='.', markersize=1,color ='gray')
        
plt.show()
Pstat=count/N
Pstat
```

```python
#стас вер
N=1000000
count=0
x=[]
y=[]
for i in range(N+1):
    x.append(i+1)
    x1 = random.uniform(0,1) 
    y1 = random.uniform(0,1)
    
    if condition(x1,y1): 
        count+=1
        
    y.append(count/(i+1))

Pstat=count/N
prob=Pstat
Pstat
```
