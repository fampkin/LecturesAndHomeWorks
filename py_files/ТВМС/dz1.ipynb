```python
import random

def coin_flip(req_symbol):
    symbols = ['orel', 'reshka']
    omega = []
    count_flips = 0
    while True:
        flip_symbol = random.choice(symbols)
        count_flips += 1
        omega.append(flip_symbol)
        if flip_symbol == req_symbol:
            return count_flips, omega

    
coin_flip('orel')

```

```python
count_m_5 = 0
N = 1_000_000

for _ in range(N):
    if coin_flip('orel')[0] < 6:
        count_m_5 += 1

print(count_m_5/N)
    
```

```python
def coin_flip_while_2_coin():
    symbols = ['orel', 'reshka']
    count_flips = 0
    omega = []
    prev_sym = ''
    while True:
        flip_symbol = random.choice(symbols)
        count_flips += 1
        omega.append(flip_symbol)
        if flip_symbol == prev_sym:
            return count_flips, omega
        prev_sym = flip_symbol
        
coin_flip_while_2_coin()
```

```python
even_count = 0
N = 100000
for _ in range(N):
    flips_count = coin_flip_while_2_coin()[0]
    if flips_count%2 == 0:
        even_count += 1
even_count/N
```

```python
def coin_flip_non_sim_while_2_drop():
    symbols = ['orel'] * 7 + ['reshka'] * 3
    count_flips = 0
    prev_sym = ''
    omega = []
    while True:
        flip_symbol = random.choice(symbols)
        count_flips += 1
        omega.append(flip_symbol)
        if flip_symbol == prev_sym:
            return count_flips, omega
        prev_sym = flip_symbol

```

```python
e2_count = 0
e4_count = 0

N = 100_000
for _ in range(N):
    flips_count = coin_flip_non_sim_while_2_drop()[0]

    if flips_count % 4 == 0:
        e2_count += 1
        e4_count += 1
    elif flips_count % 2 == 0:
        e2_count += 1

e2_count/N, e4_count/N
```
