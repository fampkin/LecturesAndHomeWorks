```python
def Omega(box, k):
    return [' '.join(comb) for comb in combinations(box, k)]

omega = Omega(box, 7)

len(omega)

```

```python
Omega1 = [0] * 8 + [1] * 9 # 0 - w;     1 - m
```

```python
count = 0
N = 1_550_000
x = []
y = []
for i in range(1, N+1):
    x.append(i)
    omega1 = list(np.random.choice(Omega1, 7, replace=False))
    if omega1.count(0) == 4:
        count += 1
    y.append(count/i)

p_hat = count/N
p_hat
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib import rcParams 

rcParams['figure.figsize'] = (30, 20)
rcParams['figure.dpi'] = 300
fig,ax = plt.subplots(figsize=(10, 5))
plt.tick_params(labelsize=16)
plt.grid(color='DarkTurquoise', alpha=0.75, linestyle=':', linewidth=1)
ax.xaxis.set_major_locator(ticker.MaxNLocator(8))
ax.xaxis.set_minor_locator(ticker.MaxNLocator(15))
plt.ylim(0.29,0.31)
plt.xlim(250,1000000)
###
plt.plot(x, y, color='b',lw=2)
plt.plot(x,[p_hat]*len(x),color='r', lw=2)
plt.show()
```
